```python
from PIL import Image
from torchvision.utils import save_image
import torchvision.transforms as transforms 
```

```python
#get an image 
img_path = "/content/HAM10000_images_part_1/ISIC_0029304.jpg"
savedir = "/content/drive/MyDrive/HAM10000/transformations_visual/"
img = Image.open(img_path)

tensor = transforms.ToTensor()
```

```python
#first transform -random rotation 
rotation = transforms.RandomRotation((45, 315))

rot_img = rotation(img)
img = tensor(rot_img)
save_image(img, savedir+"rotate.png")
```

```python
#color jitter 
jitter = transforms.ColorJitter(brightness=.3, contrast=.3)
col_img = jitter(rot_img)
img = tensor(col_img)
save_image(img, savedir+"jitter.png")
```

```python
#random solarize 
solarize = transforms.RandomSolarize(.3, 1)
sol_img = solarize(col_img)
img = tensor(sol_img)
save_image(img, savedir+"solarize.png")
```

```python
#random invert 
invert = transforms.RandomInvert(p=.3)
invert_img = invert(sol_img)
img = tensor(invert_img)
save_image(img, savedir+"invert.png")
```

```python
# sharpness 
sharp = transforms.RandomAdjustSharpness(2, p=1)
sharp_img = sharp(invert_img)
img = tensor(sharp_img)
save_image(img, savedir+"sharp.png")
```

```python
#resize 
resize = transforms.Resize(224)
re_img = resize(sharp_img)
img = tensor(re_img)
save_image(img, savedir+"resized.png")
```
