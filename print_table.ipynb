```python
from turtle import pd
import pandas as pd
data = pd.read_csv("./HAM10000_metadata")
```

```python
data.describe()
```

```python
print(data['dx'].value_counts())
```

```python
#add age group new column
def cal_age_group(age):
    try:
        age_index = int(int(age) / 10)
        return "Age " + str(age_index) + "0 - " + str(age_index + 1) + "0"
    except:
        return "Unknown"

data['age_group'] = data['age'].apply(cal_age_group)
```

```python
#now print the data and take look at age_group is valid
data
```

```python
#now group by the sex and age, dx
data_group_sex_age_dx = data.groupby(['sex', 'age_group', 'dx'])['lesion_id'].count()
data_group_sex_age_dx
```

```python
#now convert the index bkl to coulmn
data_group_sex_age_dx = data_group_sex_age_dx.unstack()
```

```python
#fill the nan to zero
data_group_sex_age_dx = data_group_sex_age_dx.fillna(0)
data_group_sex_age_dx
```

```python
data_group_sex_age =  data.groupby(['sex', 'age_group']).agg(total=('lesion_id', 'count'))
```

```python
data_group_sex_age
```

```python
#now concat the total 
result_table = pd.concat([data_group_sex_age_dx, data_group_sex_age], axis=1)
```
