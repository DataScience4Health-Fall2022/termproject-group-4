```python
md = pd.read_csv("/content/HAM10000_metadata")
```

```python
md
```

```python
dxs = md["dx"].unique()

ldxs = {}

#/content/HAM10000_segmentations_lesion_tschandl/HAM10000_segmentations_lesion_tschandl - segmented
#/content/HAM10000_images_part_2 - lesion images p2
#/content/HAM10000_images_part_1 - lesion images p1
```

```python
for dx, group in md.groupby("dx"):
    ldxs[dx] = []
    images = list(group["image_id"])
    np.random.shuffle(images)
    for i in range(100):
        ldxs[dx].append(images[i])
```

```python
ldxs
```

```python
#fp = path + fn + "_segmentation.png"
t = list(ldxs.values())[0]
path = "/content/HAM10000_segmentations_lesion_tschandl/HAM10000_segmentations_lesion_tschandl/"

imlist = []

for fn in t:
  fp = path + fn + "_segmentation.png"
  imlist.append(fp)

w,h=Img.open(imlist[0]).size
N=len(imlist)

arr=np.zeros((h,w),np.float)

for im in imlist:
    imarr=np.array(Img.open(im),dtype=np.float)
    arr=arr+imarr/N

arr=np.array(np.round(arr),dtype=np.uint8)

out=Img.fromarray(arr)
out.save("akiec_Average.png")
out.show()
```
